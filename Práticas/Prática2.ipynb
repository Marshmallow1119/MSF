a) Comece por representar os dados experimentais num gráfico
b) Calcular as somas das expressões acima
c) De seguida calcule o declive, a ordenada na origem e o coeficiente de determinação ou de correlação 

```python
x=np.array([222.0,207.5,194.0,171.5,153.0,133.0,113.0,92.0])
y=np.array([2.3,2.2,2.0,1.8,1.6,1.4,1.2,1.0])


def regreslinear(x,y): #l(cm) e x(cm)

    somx=np.sum(x);
    somy=np.sum(y);
    somxy=np.sum(x*y);
    somx2=np.sum(x**2);
    somy2=np.sum(y**2);
    N=len(x);
    m=((N*somxy)-(somx*somy))/((N*somx2)-(somx)**2)

    b=((somx2*somy)-(somx*somxy))/((N*somx2)-(somx)**2)

    r2=(((N*somxy)-(somx*somy))**2)/(((N*somx2)-((somx)**2))*((N*somy2)-((somy)**2)))

    delM=np.abs(m)*(np.sqrt(((1/r2)-1)/(N-2)))
    delB=delM*(np.sqrt(somx2/N))

    
    return somx,somy,somxy,somx2,somy2,N,m,b,r2,delM,delB;

print(regreslinear(x,y))


plt.plot(x,y)
plt.suptitle('REGRESSÃO LINEAR')
plt.xlabel("L(cm)")
plt.ylabel("X(cm)")
plt.grid()
plt.show()
```

d) faça um gráfico com os pontos experimentais e a reta cujos parâmetros m e b calculou anteriormente

```python
plt.scatter(x,y,marker='x',color="red")
somx,somy,somxy,somx2,somy2,N,m,b,r2,delM,delB=regreslinear(x,y)
retaajustada=m*x+b
plt.plot(x, retaajustada)
plt.xlabel("L(cm)")
plt.ylabel("X(cm)")
plt.grid()
plt.show()
```

Foi medida a energia por segundo (potência) emitida por um corpo negro (corpo que absorve toda a energia que incide 
nele) de área 100 cm2 em função da temperatura absoluta, �, e registada na seguinte tabela



T (K) 200. 300. 400. 500. 600. 700. 800. 900. 1000. 1100.

E (J) 0.6950 4.363 15.53 38.74 75.08 125.2 257.9 344.1 557.4 690.7


a) Apresente estas medições num gráfico. A analisar o gráfico, a relação entre a energia emitida e a temperatura é linear?


```python
temperatura=np.array([200.0,300.0,400.0,500.0,600.0,700.0,800.0,900.0,1000.0,1100.0])
energia=np.array([0.6950,4.363,15.53,38.74,75.08,125.2,257.9,344.,557.4,690.7])

somx,somy,somxy,somx2,somy2,N,m,b,r2,delM,delB=regreslinear(temperatura,energia)

plt.scatter(temperatura,energia,marker="x")
retaajustada2=m*temperatura+b
plt.plot(temperatura,retaajustada2)
plt.xlabel("temperatura(K)")
plt.ylabel("energia(J)")
plt.show()

```

Não é linear

b) Apresente as medições num gráfico log-log. Qual a dependência entre as quantidade energia emitida e a temperatura?


```python
plt.plot(np.log(temperatura),np.log(energia),"o")
a = np.polyfit(np.log(temperatura), np.log(energia), 1)

t = np.linspace(np.log(temperatura[0])*0.9, np.log(temperatura[-1])*1.1, 100)
plt.plot(t, a[0]*t+a[1])

plt.show()
```

Problema cap 1.9 Regressão linear pelo método dos mínimos quadráticos



Foi medida a atividade de uma amostra do isótopo radioativo 131I tem de 5 em 5 dias. Os valores medidos da atividade com 
o tempo são, em mCi:



9.676 , 6.355, 4.261, 2.729, 1.862, 1.184, 0.7680, 0.4883, 0.3461, 0.2119



a) Apresente estas medições num gráfico. A analisar o gráfico, a relação entre a atividade e o tempo é linear?


```python
tempodias=np.array([5,10,15,20,25,30,35,40,45,50]) # ou podemos colocar np.arrange([0,50,5])
atividade=np.array([9.676 , 6.355, 4.261, 2.729, 1.862, 1.184, 0.7680, 0.4883, 0.3461, 0.2119])

somx,somy,somxy,somx2,somy2,N,m,b,r2,delM,delB=regreslinear(tempodias,atividade)

plt.scatter(tempodias,atividade)
retaajustada3=m*tempodias+b
plt.plot(tempodias,retaajustada3)
plt.xlabel("tempodias(K)")
plt.ylabel("atividade(J)")
plt.show()
```

É linear

b) Apresente as medições num gráfico semilog. Como depende a atividade com o tempo?
