```python
import matplotlib.pyplot as plt
(x0,y0)=(0,0)
(x,y)=(3,4)
(x1,y1)=(4,-3)
plt.arrow(x0,y0,x,y,color='r',width=0.1)
plt.arrow(x0,y0,x1,y1,color='r',width=0.1)
plt.axis('equal')
```
