12. Determine a evolução temporal da velocidade de um ciclista, se este produzir con6nuamente a potência 0.4 cv e par6r com um empurrão de 1 m/s?
a) Qual a sua velocidade terminal?


```python
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt


def metododeeuler(dt,t0,tf,v0,g,P0_cic,ang,A,Cres,M,Par,Coe):

    #dt passo de tempo delta t(apontamentos caderno)

    #inicialização
    n=int((tf-t0)/dt+0.1) # +0.1 para garantir não arredondar para baixo
    t=np.zeros(n+1) # n+1 elementos; último índice n
    y=np.zeros(n+1)
    vx=np.zeros(n+1)
    ax=np.zeros(n+1)
    x=np.zeros(n+1)
    a=np.zeros(n+1)
    p0_cic=np.zeros(n+1)
    angulo=np.zeros(n+1)
    cres=np.zeros(n+1)
    m=np.zeros(n+1)
    par=np.zeros(n+1)
    coe=np.zeros(n+1)

    ax[0]=g
    vx[0]=v0
    t[0]=t0
    a[0]=A
    p0_cic[0]=P0_cic 
    cres[0]=Cres
    angulo[0]=ang
    m[0]=M
    par[0]=Par
    coe[0]=Coe

    # Método de Euler (n+1 elementos)
    for i in range(n):
        ax[i] = (p0_cic[0]/(m[0]*vx[i]))-((cres[0]/(2*m[0]))*a[0]*par[0]*vx[i]*vx[i])-(coe[0]*9.8)
        x[i+1]=x[i]+vx[i]*dt
        vx[i+1]=vx[i] + ax[i]*dt # atualizar velocidade sabendo aceleração
        t[i+1]=t[i]+dt

    return n,ax,x,vx,t

#paraámetros
P0_cic=0.4*736 #multiplicar 0.4 por 736 para colocar em w
n,ax,x,vx,t=metododeeuler(0.001,0,700,1,9.8,P0_cic,0,0.30,0.9,75,1.225,0.004)


plt.plot(t,vx)
print(vx)
plt.xlabel("tempo")
plt.ylabel("Velocidade")
plt.show()

for i in range(n):
    if ax[i+1]<0+0.001 and ax[i+1]>0-0.001:
        print(f"A velocidade é: {vx[i+1]} m/s \n")
        break
```

11. Muitos ciclistas amadores conduzem com o tronco levantado. Esta posição aumenta a área de varrimento para o dobro. 



a) Calcule a potência a desenvolver pelo ciclista se a velocidade for 30 km/h. 




```python
v = 30/3.6
P = ((0.9/2) * 0.6*1.225*v**2 + 0.004*75*9.80) * v

print(f"Potência a desenvolver: {P:.2f} w\n")
```
