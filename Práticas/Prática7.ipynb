a) A rota√ß√£o √© nula.


```python
def euler():
    dt = 0.01
    tf = 2
    n = int(tf/dt+0.1)

    t = np.empty(n)
    vy = np.empty(n)
    ay = np.empty(n)
    y = np.empty(n)
    x = np.empty(n)
    vx = np.empty(n)
    ax = np.empty(n)
    z = np.empty(n)
    vz = np.empty(n)
    az = np.empty(n)
    w = np.array([0, 0, 0])#rota√ß√£o da bola
    v = np.array([35.56,6.27 ,0 ])#velocidade da bola decomposicao de forcas ou seja vcos,vsen,0
    r = np.array([-10, 1, 0])#posicao inicial

    g = 9.80
    vt = 100/3.6
    vy[0] = v[1]
    t[0] = 0
    y[0] = r[1]
    x[0] = r[0]
    z[0] = r[2]
    vx[0] = v[0]
    vz[0] = v[2]

    for i in range(n-1):
        F = 0.5 * np.pi * (0.0335**2) * 1.225 * 0.0335 * np.cross(w, np.array([vx[i], vy[i], vz[i]]))
        a = F/0.057
        t[i+1] = t[i]+dt
        vv = np.abs(np.sqrt(vx[i]**2 + vy[i]**2 + vz[i]**2))
        D = g/vt**2
        ax[i] = -D*vv*vx[i]+a[0]
        ay[i] = -g-D*vv*vy[i]+a[1]
        az[i] = -D*vv*vz[i]+a[2]
        vy[i+1] = vy[i]+ay[i]*dt
        vx[i+1] = vx[i]+ax[i]*dt
        vz[i+1] = vz[i]+az[i]*dt
        x[i+1] = x[i]+vx[i]*dt
        y[i+1] = y[i]+vy[i]*dt
        z[i+1] = z[i]+vz[i]*dt

    return t, y, vy, x, vx, z, vz, n


t, y, vy, x, vx, z, vz, n = euler()

for i in range(n-1):
    if vy[i+1] < 0+0.05 and vy[i+1] > 0-0.05:
        print(f'Altura m√°xima - tempo: {t[i+1]}; posi√ß√£oy: {y[i+1]};\n')

for i in range(n-1):
    if y[i+1] < 0+0.03 and y[i+1] > 0-0.03:
        print(f'Chegada ao solo - tempo: {t[i+1]}; posi√ß√£ox:¬†{x[i+1]};\n')
```

b) Calcule numericamente a lei da velocidade usando o m√©todo de Euler, e compare com o resultado anal√≠tico. 
Considere nula a velocidade inicial e a posi√ß√£o inicial 4 m.


c) Calcule numericamente a lei do movimento nas condi√ß√µes da al√≠nea anterior. Compare com o resultado anal√≠tico.



```python

def metododeeuler(dt,t0,tf,vx0,x0,w):

    #dt passo de tempo delta t(apontamentos caderno)

    #inicializa√ß√£o
    n=int((tf-t0)/dt+0.1) # +0.1 para garantir n√£o arredondar para baixo
    t=np.zeros(n+1) # n+1 elementos; √∫ltimo √≠ndice n
    vx=np.zeros(n+1)
    ax=np.zeros(n+1)
    x=np.zeros(n+1)

    g=9.8

    
    vx[0]=vx0
    t[0]=t0
    x[0]=x0


    # M√©todo de Euler (n+1 elementos)
    for i in range(n):
        ax[i]=-w**2*x[i]
        x[i+1]=x[i]+vx[i]*dt
        vx[i+1]=vx[i] + ax[i]*dt # atualizar velocidade sabendo acelera√ß√£o
        t[i+1]=t[i]+dt

    return ax,x,vx,t


dt=0.01
A=4
k=1
m=1
tf=100
t0=0

w=math.sqrt(k/m)
ax,x,vx,t=metododeeuler(dt,t0,tf,0,A,w)


phi=0
t_analitico=np.linspace(0,tf,1000)
x_analitico=A*np.cos(w*t_analitico+phi)
v_analitico=-A*np.sin(w*t_analitico+phi)

plt.plot(t,x,label="Euler")
plt.plot(t_analitico,x_analitico,label="Anal√≠tico")
plt.legend()
plt.xlabel("Tempo(s)")
plt.ylabel("Posi√ß√£o(m)")
plt.title('Movimento Harm√≥nico simples:Posi√ß√£o vs tempo')
plt.show()

plt.plot(t,vx,label="Euler")
plt.plot(t_analitico,v_analitico,label="Anal√≠tico")
plt.legend()
plt.xlabel("Tempo(s)")
plt.ylabel("Velociddde(m/s)")
plt.title('Movimento Harm√≥nico simples:velocidade vs tempo')
plt.show()

```

```python
def metododeeuler_cromer(dt,t0,tf,vx0,x0,w):
    #movimento harmonico simples
    #integracao numerica de dx/dt =vx pelo metodo de euler
    #dt passo de tempo delta t(apontamentos caderno)


    #a=-k/m *x0 com a acelaracao, k constante mola, m massa e c posicao inicial
    #Esta funcao entra com w= sqrt(k/m)
    #Logo, a=-w*2 *x0

    #dt=0.5
    #tf=4
    #t0=v0x=x0=0

    #inicializa√ß√£o
    n=int((tf-t0)/dt+0.1) # +0.1 para garantir n√£o arredondar para baixo
    t=np.zeros(n+1) # n+1 elementos; √∫ltimo √≠ndice n
    vx=np.zeros(n+1)
    ax=np.zeros(n+1)
    x=np.zeros(n+1)


    
    vx[0]=vx0
    t[0]=t0
    x[0]=x0

    
    # M√©todo de Euler (n+1 elementos)
    for i in range(n):
        t[i+1]=t[i]+dt   
        ax[i]=-w**2*x[i] 
        vx[i+1]=vx[i] + ax[i]*dt # atualizar velocidade sabendo acelera√ß√£o  
        x[i+1]=x[i]+vx[i+1]*dt
                # atualizar o tempo
    return ax,x,vx,t


dt=0.01
A=4
k=1
m=1
tf=100
t0=0
w=math.sqrt(k/m)

ax,x,vx,t=metododeeuler_cromer(dt,t0,tf,0,A,w)


phi=0
t_analitico=np.linspace(0,tf,100)
x_analitico=A*np.cos(w*t_analitico+phi)
v_analitico=-A*np.sin(w*t_analitico+phi)

plt.plot(t,x,label="Euler-Cromer")
plt.plot(t_analitico,x_analitico,label="Analitico")
plt.legend()
plt.xlabel("Tempo(s)")
plt.ylabel("Posi√ß√£o(m)")
plt.title("Movimento Harm√≥nico simples:Posi√ß√£o vs tempo")
plt.show()

plt.plot(t,vx,label="Euler-Cromer")
plt.plot(t_analitico,v_analitico,label="Analitico")
plt.legend()
plt.xlabel("Tempo(s)")
plt.ylabel("Velociddde(m)")
plt.title("Movimento Harm√≥nico simples:velocidade vs tempo")
plt.show()

```
