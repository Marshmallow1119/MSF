b) Construa um programa que determine a velocidade do objeto, usando o método de 
Euler, no intervalo de tempo 0, 4 s . Qual a velocidade em 3s?

```python

def metododeeuler(dt,t0,tf,y0,vy0,g):

    #dt passo de tempo delta t(apontamentos caderno)

    #inicialização
    n=int((tf-t0)/dt+0.1) # +0.1 para garantir não arredondar para baixo
    t=np.zeros(n+1) # n+1 elementos; último índice n
    y=np.zeros(n+1)
    vy=np.zeros(n+1)
    ay=np.zeros(n+1)
    x=np.zeros(n+1)

    
    vy[0]=vy0
    t[0]=t0
    y[0]=y0



    # Método de Euler (n+1 elementos)
    for i in range(n):
        ay[i] = g # queda livre 
                # (em geral pode ser qualquer função de x[i] e vx[i])
        y[i+1]=y[i]+vy[i]*dt
        vy[i+1]=vy[i] + ay[i]*dt # atualizar velocidade sabendo aceleração
        t[i+1]=t[i]+dt

    return ay,y,vy,t



#paraámetros

ay,y,vy,t=metododeeuler(1,0,4.0,0,0,9.8)
print("A velocidade é",vy[3])
plt.plot(t,vy)
plt.xlim([0,4])
plt.ylim([0,40])
plt.xlabel("tempo")
plt.ylabel("Velocidade")
plt.show()
```

d) Resolva as alíneas anteriores, considerando a resistência do ar. Resolva usando 
o método de Euler. A velocidade terminal da bola no ar é de 100 km/h.

```python
import numpy as np
import matplotlib.pyplot as plt


dt = 0.01
tf = 2.0
n = int(tf/dt+0.1)

t = np.empty(n)
vy = np.empty(n)
ay = np.empty(n)
y = np.empty(n)

g = 9.80
vt = 100/3.6
vy0 = 10
vy[0] = vy0
t[0] = 0
y[0] = 0

for i in range(n-1):  # Calular as equações de acordo com a resistencia do ar
    t[i+1] = t[i]+dt
    vv = np.abs(vy[i])
    dres = g/vt**2
    ay[i] = -g-dres*vv*vy[i]  # formula no formulario
    vy[i+1] = vy[i]+ay[i]*dt
    y[i+1] = y[i]+vy[i]*dt


for i in range(n-1):  # A altura é maxima quando v=0
    if vy[i] > 0-dt and vy[i+1] < 0+dt:
        print(
            f'altura máxima - tempo: {t[i+1]}; posição: {y[i+1]}; velocidade (m/s): {vy[i+1]};')

for i in range(n-1):  # Chega ao solo quando y=0
    if y[i] < 0+dt and y[i+1] > 0-dt and i > 3:
        print(
            f'Chegada ao solo - tempo: {t[i+1]}; posição: {y[i+1]}; velocidade (m/s): {vy[i+1]};')


plt.plot(t, y)
plt.xlabel('Tempo (s)')
plt.ylabel('Posição (m)')
plt.title('Queda da bola com resistencia do ar')
plt.show()
```
